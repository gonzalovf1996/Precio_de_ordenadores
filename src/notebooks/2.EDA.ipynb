```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd

# Gráficos
# ==============================================================================
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('default')
```

# EDA
---------------------------

```python
train = pd.read_csv('../data/processed/train_tratado.csv')
```

### Análisis de la correlación

```python
sns.distplot(train.Price_euros);
```

```python
plt.figure(figsize=(20,20))
sns.set_style('whitegrid')
sns.heatmap(train.corr(), annot=True);
```

### Observaciones:

Se aprecia multicolinearidad en dos columnas ('2 in 1 Convertible' & 'Touchscreen' ). Por ende, como '2 in 1 Convertible' tiene menor correlación con nuestra target, será exclusida del análisis.
